```python
from combine_data import *
import os
import pandas as pd
```

```python
# Get list of all folders, excluding hidden folders
folders = [item for item in os.listdir('.') if os.path.isdir(item) and not ((item.startswith('.')) or (item.startswith('_')))]

print(folders)
```

```python
file_names = {}
for exp in folders:
    csv_files = [file for file in os.listdir(exp) if file.endswith('.csv')]
    file_names[exp] = csv_files

print(file_names['Exp 1A'])
print(len(file_names['Exp 1A']))
```

```python
results = {'Sub_ID': [],
           'Exp': [],
           'Acc_LV': [],
           'RT_LV': [],
           'Acc_HV': [],
           'RT_HV': []}

for exp, files in file_names.items():
    for i in files:
        sub_id = i[:-4] #make the sub id the file name without the '.csv'
        #low reward = 1, high reward = 2
        #valid = 1, neutral = 2

        sub_df = open_file(exp, i)

        #low, valid
        sub_temp = isolate_column_by_value(sub_df, exp, 1, 1)
        acc_lv = get_average(sub_temp, exp)
        rt_lv = get_rt(sub_temp, exp)

        #low, neutral
        sub_temp = isolate_column_by_value(sub_df, exp, 1, 2)
        acc_ln = get_average(sub_temp, exp)
        rt_ln = get_rt(sub_temp, exp)

        #high, valid
        sub_temp = isolate_column_by_value(sub_df, exp, 2, 1)
        acc_hv = get_average(sub_temp, exp)
        rt_hv = get_rt(sub_temp, exp)

        #high, neutral
        sub_temp = isolate_column_by_value(sub_df, exp, 2, 2)
        acc_hn = get_average(sub_temp, exp)
        rt_hn = get_rt(sub_temp, exp)

        #append to results dict
        results['Sub_ID'].append(sub_id)
        results['Exp'].append(exp)
        results['Acc_LV'].append(acc_lv)
        results['RT_LV'].append(rt_lv)
        results['Acc_HV'].append(acc_hv)
        results['RT_HV'].append(rt_hv)

results_df = pd.DataFrame(results)
results_df.to_csv('Across_Studies_Analysis.csv', index=False)
```
